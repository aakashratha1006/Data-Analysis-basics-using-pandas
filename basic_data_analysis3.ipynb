```python
import pandas as pd
movies = pd.read_csv('http://bit.ly/imdbratings')
```

```python
movies.head()
```

```python
movies.shape
```

```python
booleans = []
for length in movies.duration:
    if length >= 200:
        booleans.append(True)
    else:
        booleans.append(False)
```

```python
booleans[0:5]
```

```python
len(booleans)
```

```python
is_long = pd.Series(booleans)
```

```python
is_long.head()
```

```python
movies[is_long]
```

```python
# Above one was longer method
is_long = movies.duration >= 200
is_long.head()
```

```python
# or
movies[movies.duration >= 200]
```

```python
movies[movies.duration >= 200].genre
```

```python
movies.loc[movies.duration >= 200, 'genre'] # Better Pracice
```
